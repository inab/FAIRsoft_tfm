```python
from pandas import DataFrame
from pandas import read_csv
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
canonFAIR = read_csv('scores/FAIRcanon.txt', sep='\t')
```

```python
canonFAIR
```

```python
# Basic correlogram
sns.pairplot(canonFAIR)
```

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = canonFAIR.corr()

sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
plt.title('Correlation between the FAIRsoft scores, number of sources and number of types of canonical tools')
```
